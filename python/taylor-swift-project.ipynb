```python
# Change release date to release year and month

swift['release_date'] = pd.to_datetime(swift['release_date'])
swift['release_month'] = swift['release_date'].dt.month

swift['release_year'] = swift['release_date'].dt.year
```

```python
# Convert duration from milliseconds to minutes
swift['duration_min'] = swift['duration_ms'] / (1000 * 60)
```

```python
# And we drop the release date and duration ms columns
swift = swift.drop(['release_date', 'duration_ms'], axis=1)
```
