```python



plt.style.use('_mpl-gallery')
 
# code to plot 

plt.figure(figsize=(6,6))


plt.scatter(x = iris_df['SepalLengthCm'],y=iris_df['PetalLengthCm'], color = 'pink')


# title of the plot

plt.title("A scatter plot for Speal length and Petal length")

# add labels to the axis

plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')



plt.show()
 

```

```python
plt.style.use('_mpl-gallery')

# Create the figure
plt.figure(figsize=(6,6))

# Define colors for each species
colors = {'Iris-setosa': '#40e0d0', 'Iris-versicolor': '#ff6347', 'Iris-virginica': '#9370db'}

# Plot each species
for Species, color in colors.items():
    subset = iris_df[iris_df['Species'] == Species]
    plt.scatter(subset['SepalLengthCm'], subset['PetalLengthCm'], color=color, label=Species)

# Title of the plot
plt.title("A scatter plot for Sepal length and Petal length")

# Add labels to the axes
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
```

```python
species_count = iris_df['Species'].value_counts()

# Plotting
plt.style.use('_mpl-gallery')
plt.figure(figsize=(6,3))

# Create the bar plot
plt.bar(species_count.index, species_count.values,width = 0.3, color='red', label='Species count')

# Title of the plot
plt.title("A bar plot to count the different species")

# Add labels to the axis
plt.xlabel("Species")
plt.ylabel('Count of the Species')



# Show the plot
plt.show()
```

```python


plt.style.use('_mpl-gallery')

plt.figure(figsize=(6,3))
plt.bar(iris_df['Species'], iris_df['SepalLengthCm'], width = 0.3, color="teal")
plt.title("Median of Sepal Length in cm by Species")
plt.xlabel("Species name")
plt.ylabel("Sepal length (cm)")
plt.legend()

plt.show()

plt.clf()
```

```python
plt.style.use('_mpl-gallery')

# Create the figure
plt.figure(figsize=(6, 6))

# Plot the histogram for Sepal Length
plt.hist(iris_df['SepalLengthCm'], width = 0.4, color='green', edgecolor='black')

# Title of the plot
plt.title("Histogram of Sepal Length")

# Add labels to the axes
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python
data_plot_group=iris_df.groupby('Species').agg({'SepalLengthCm':'mean','SepalWidthCm':'mean',
                                           'PetalLengthCm':'mean','PetalWidthCm':'mean'}).reset_index()
data_plot_group
```

```python
plt.style.use('_mpl-gallery')
 
# code to plot 

plt.figure(figsize=(6,3))

plt.bar(data_plot_group.Species,data_plot_group.SepalLengthCm, color='#40e0d0',width = 0.3,label='Species')

# title of the plot

plt.title("A bar plot with average Sepal Length per species",fontsize=16,color='Green')

# add labels to the axis

plt.xlabel("Species",fontsize=10)

plt.xticks(rotation=0, fontsize=10)

plt.ylabel('Average sepal length',fontsize=10)

plt.yticks(fontsize=10)

# grid

# legend

plt.legend()

plt.show()
 
```

### use of extended librraies for plotting data 
- Seaborn

```python
import seaborn as sns
```

```python
plt.style.use('_mpl-gallery')
plt.figure(figsize=(6,4))
sns.scatterplot(data = iris_df, x='SepalLengthCm', y= 'SepalWidthCm',hue = 'Species')
plt.title("A seaborn scatterplot with average Sepal Length per species",fontsize=14,color='#61d290')
# add labels to the axis
plt.xlabel("SepalLengthCm",fontsize=12)
plt.ylabel('SepalWidthCm',fontsize=12)

plt.legend()

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(data=iris_df,x='SepalLengthCm',y='SepalWidthCm', hue='Species',palette='deep')

# title of the plot

plt.title("A seaborn scatterplot with average Sepal Length per species",fontsize=14,color='#61d290')

# add labels to the axis

plt.xlabel("SepalLengthCm",fontsize=12)

plt.ylabel('SepalWidthCm',fontsize=12)

plt.legend()

plt.show()
 
```
