**1. Introduction to Pandas**

Pandas is an open-source library providing high-performance, easy-to-use data structures and data analysis tools for Python. It is built on top of NumPy and integrates well with other Python libraries like SciPy, Matplotlib, and scikit-learn, making it a cornerstone of the data science ecosystem.

### **Key Features:**

- **Data Structures:** Provides Series and DataFrame for handling one-dimensional and two-dimensional data, respectively.
- **Data Manipulation:** Offers functions for cleaning, transforming, and analyzing data.
- **Performance:** Built on top of NumPy, ensuring fast and efficient numerical operations.
- **Integration:** Seamlessly integrates with other popular Python libraries for data science and machine learning.

#### join command using new datasets called left and right

````python
left = pd.DataFrame(

    {

        "A": ["A0", "A1", "A2", "A3"],

        "B": ["B0", "B1", "B2", "B3"],

        "key": ["K0", "K1", "K0", "K1"],

    })
 
right = pd.DataFrame({"C": ["C0", "C1"], "D": ["D0", "D1"]}, index=["K0", "K1"])
 
````

````python
join_df = left.join(right, on = 'key')
join_df
````
