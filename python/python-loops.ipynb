#### Leap Year Checker

```python
year = eval(input("please enter a Year"))
if year%4 == 0 or year%400 == 0:
    print("Its a leap Year")
elif year%100 != 0:
    print("Its a normal year")
else:
    print("Please enter a positive 4 digit number")
```

#### Vowel or consonant
##### Method 1

```python
letter = input("please enter an english alphabet")
if letter == 'a' or letter == 'e' or letter == 'i' or letter == 'o' or letter == 'u': 
        print("Vowel") 
else: 
        print("Consonant") 
  
```

##### Method 2

```python
l = input("Enter the character: ")

if l.lower() in ('a', 'e', 'i', 'o', 'u'):

  print("Vowel")

elif l.upper() in ('A', 'E', 'I', 'O', 'U'):

  print("Vowel")

else:

  print("Consonant")

```

#### Simple calculator for basic operations  using two numbers 

```python
x = eval(input("Enter first number"))
y= eval(input("Enter second number"))
operator = input("which operation you want choose from +, -,*, /")
if operator == '+':
    print(x+y)
elif operator == '-':
    print(x-y)
elif operator == '*':
    print(x*y)
else:
    print(x/y)
    


```
